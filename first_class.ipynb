```python
import numpy as np
mat1 = np.array([[1,3,4],[1,7,4]])
mat2 = np.array([[8,3,9],[2,5,6]])
result = np.add(mat1, mat2)
print(result)
```

```python
import numpy as np
mat2 = np.transpose([[8,3,9],[2,5,6]])
print(mat2)
```

```python
import numpy as np
mat1 = np.array([[1,2], [5,6]])
mat2 = np.array([[1,2], [5,6]]) 
result = np.dot(mat1, mat2)
print(result)
```
